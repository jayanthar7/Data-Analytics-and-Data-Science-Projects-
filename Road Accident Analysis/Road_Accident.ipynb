```python
current_year = df['year'].max()
current_year
previous_year = df['year'].min()
previous_year
cy_casualties = df[df['year'] == current_year]['number_of_casualties'].sum()
py_casualties = df[df['year'] == previous_year]['number_of_casualties'].sum()
print(f"Total Current Year Casualties : {cy_casualties}")
print(f"Total Previous Year Casualties: {py_casualties}")
```

```python
cy_fatal_casualties = df[(df['year'] == current_year) & 
                      (df['accident_severity'].str.lower() == 'fatal')]['number_of_casualties'].sum()
cy_serious_casualties = df[(df['year'] == current_year) & 
                      (df['accident_severity'].str.lower() == 'serious')]['number_of_casualties'].sum()
cy_slight_casualties = df[(df['year'] == current_year) & 
                      (df['accident_severity'].str.lower() == 'slight')]['number_of_casualties'].sum()
print(f"Current Year Fatal Casualties: {cy_fatal_casualties}")
print(f"Current Year Serious Casualties: {cy_serious_casualties}")
print(f"Current Year Slight Casualties: {cy_slight_casualties}")
```

```python
py_fatal_casualties = df[(df['year'] == previous_year) & 
                      (df['accident_severity'].str.lower() == 'fatal')]['number_of_casualties'].sum()
py_serious_casualties = df[(df['year'] == previous_year) & 
                      (df['accident_severity'].str.lower() == 'serious')]['number_of_casualties'].sum()
py_slight_casualties = df[(df['year'] == previous_year) & 
                      (df['accident_severity'].str.lower() == 'slight')]['number_of_casualties'].sum()
print(f"Previous Year Fatal Casualties: {py_fatal_casualties}")
print(f"Previous Year Serious Casualties: {py_serious_casualties}")
print(f"Previous Year Slight Casualties: {py_slight_casualties}")
```

```python
data = {
    'current year fatal casualities': cy_fatal_casualties,
    'current year slight casualities': cy_slight_casualties,
    'current year serious casualities': cy_serious_casualties
}
df = pd.DataFrame(data = data, index = ['0'])
df
```

```python
data = {
    'category': ['current year fatal casualties',
                 'current year slight casualties',
                 'current year serious casualties'
                 ],
    'count': [cy_fatal_casualties, cy_slight_casualties, cy_serious_casualties]
}
df = pd.DataFrame(data)
df
```

```python
percentage = (
    df.groupby('accident_severity')['number_of_casualties'].sum() / df['number_of_casualties'].sum()) * 100
percentage.sort_values(ascending = False).round(2).to_frame().astype(str) + ' %'
```

```python
vehicles_casualties = (
    df.groupby('vehicle_type')['number_of_casualties'].sum()
    .reset_index()
)
vehicles_casualties
```

```python
def vehicle(v):
    if any (x in v for x in ['Taxi/Private hire car', 'Car']):
        return 'Car'
    elif any (x in v for x in ['Motorcycle 125cc and under', 'Motorcycle 50cc and under', 'Motorcycle over 125cc and up to 500cc', 'Motorcycle over 500cc']):
        return 'Bike'
    elif any (x in v for x in ['Bus or coach (17 or more pass seats)', 'Minibus (8 - 16 passenger seats)']):
        return 'Bus'
    elif any (x in v for x in ['Goods 7.5 tonnes mgw and over', 'Goods over 3.5t. and under 7.5t', 'Van / Goods 3.5 tonnes mgw or under']):
        return 'Van'
    elif any (x in v for x in ['Agricultural vehicle']):
        return 'Agricultural Vehilce'
    else:
        return 'Other Vehicles'
df['vehicle_group'] = df['vehicle_type'].map(vehicle)
```

```python
vehicle_casualties = (
    df.groupby('vehicle_group')['number_of_casualties'].sum()
)
vehicle_casualties.sort_values(ascending = False).to_frame()
```

```python
monthty_trend = df[df['accident_date'].dt.year == 2022].groupby(
    df['accident_date'].dt.month_name()
)['number_of_casualties'].sum()

monthty_trend.sort_values(ascending = False).to_frame()
```

```python
road_type = (
    df.groupby('road_type')['number_of_casualties'].sum().sort_values(ascending = False)
)
road_type.to_frame()
```

```python
area_total = (
    df.groupby('urban_or_rural_area')['number_of_casualties'].sum()
)
area_percentage = round((area_total / area_total.sum()) * 100, 2).astype(str) + ' %'
df_area = pd.concat([area_total, area_percentage], axis = 1)
df_area
```

```python
df.light_conditions.value_counts()
```

```python
def light_mapping(l):
    if any(x in l for x in['Daylight']):
        return "Day"
    else:
        return "Night"


df['light_cond'] = df['light_conditions'].map(light_mapping)
```

```python
light_cond =(
    df.groupby('light_cond')['number_of_casualties'].count()
    .sort_values(ascending = False).astype(str) + ' - Nos of People Has Been Affected'
)
light_cond.to_frame("Count of Casualties")
```

```python
light_cond = (
    df.groupby('light_cond')['number_of_casualties'].sum()
)
percentage = round((light_cond / light_cond.sum()) * 100, 2).astype(str) + ' %'
df_light = pd.concat([light_cond, percentage], axis = 1,
                    keys = ['Total Nos of Casualties', 'Percentage %'])
df_light
```
